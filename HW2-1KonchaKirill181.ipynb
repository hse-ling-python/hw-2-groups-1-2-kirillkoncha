## Задание3

С помощью nltk токенизируем текст, убираем из него стопслова и знаки препинания, приводим все к нижнему регистру

```python
%%time
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
sw = stopwords.words('russian')
words = [w.lower() for w in word_tokenize(text) if w.isalpha()]
filtered = [w for w in words if w not in sw]
filtered[:20]
```

Делаем разбор с помощью pymorphy и сохраняем список, где каждый элемент -- это строка вида лемма + часть речи. Для того, чтобы все прошло гладко сначала очищаем с помощью регулярного выражения слова от знаков препинания. 

Так как порядок слов должен быть именно как в тексте, то все результаты записываем в список, а не словарь.

```python
%%time
import re
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
list_of_POS = [] #задел на задание 4
lemmas_POS = []

for word in text.split():
    word = re.sub(r'[^\w\s]','',word)
    ana = morph.parse(word)
    if ana[0].tag.POS:
        string = str(ana[0].normal_form) + ': ' + str(ana[0].tag.POS)
        lemmas_POS.append(string)
        list_of_POS.append(ana[0].tag.POS)

lemmas_POS[:20]
```

Cохраняем все это в json

```python
with open('pymorphy2.json', 'w') as f:
    json.dump(lemmas_POS, f, ensure_ascii=False)
```

## Задание 4
Сначала сделаем список из того, какие части речи есть в тексте. Так как у нас уже есть список из одних только перечеслений частей речи, то мы легко можем с помощью Counter посчитать сколько чего есть. 


```python
import collections
POS_totall = []
c = collections.Counter()

for word in list_of_POS:
    if word not in POS_totall:
        POS_totall.append(word)

for word in list_of_POS:
    c[word] += 1

for word in POS_totall:
    c[word] = c[word] / len(list_of_POS) * 100
    c[word] = str(c[word]) + ' ' + '%'

c
```
